```python
linear= lambda x: x-0.5
quadratic= lambda x: (x-0.25)*(x-0.75)
cubic= lambda x: (x-0.33)*(x-0.5)*(x-0.66)

funciones= [linear,quadratic,cubic]

x=np.linspace(0,1, num=200)

fig, axs= plt.subplots(1,3, figsize=(10,4))
for f,ax in zip(funciones, axs.flatten()):
    ax.plot(x, f(x))
    
plt.show
```

```python
rng=default_rng(24)

fig, axs= plt.subplots(1,3, figsize=(10,4))
for f,ax in zip(funciones, axs.flatten()):
    rng.shuffle(x)
    ax.plot(x, f(x))
    
plt.show()
```

```python
numeros_clase=pd.DataFrame(
    {'numero':[86,28,4,55,108,96,73,13,97,44,25,19,32],
    'nombre':['gustavo','erik','jose','eduardo','jose eduardo','natalia','agustin','pau','loreli','nelther','caludia','ivan','jen']
    })
```

```python
x=np.arange(len(numeros_clase['nombre']))
fig,ax=plt.subplots(figsize=(15,4))
ax.bar(x,numeros_clase['numero'], align='edge')
ax.set_xticks(x)
ax.set_xticklabels(numeros_clase['nombre'])

plt.show()
```

```python
rng=default_rng(25)

c_puntos1=(-1,1)
size1=50

c_puntos2=(1,-1)
size=50

corte=0
pend=1
rango=(-4,4)

fig,ax=plt.subplots()

x1=rng.normal(loc=c_puntos1[0], size=size)
y1=rng.normal(loc=c_puntos1[1], size=size)

x2=rng.normal(loc=c_puntos2[0], size=size)
y2=rng.normal(loc=c_puntos2[1], size=size)

ax.scatter(x1,y1, color='green')
ax.scatter(x2,y2, color='red')

x_linspace=np.linspace(rango[0],rango[1], num=20)
ax.plot(x_linspace,pend*x_linspace+corte, linewidth=25)

plt.show()
```

```python
# Doble eje
x=np.arange(1,11)

fig,ax=plt.subplots()

ax.plot(x,np.exp(x))
ax.set_ylabel('exp')
ax.set_xlabel('x')

ax2=ax.twinx() #config eje 2
ax2.plot(x,np.log(x),'bo-')
ax2.set_ylabel('log')
ax.set_title('Exp vs log de x')

ax.legend(labels=('exp','log'))


plt.show()
```
