```python
rng=default_rng()
size=5000
x=rng.normal(size=size)
y=rng.normal(size=size)

fig, ax=plt.subplots()
ax.scatter(x,y,alpha=0.3)
plt.show
```

```python
rng=default_rng(18)
size=100
x=rng.uniform(size=size)
y=1.0*x+rng.normal(scale=0.2,size=size)

fig,ax=plt.subplots()
ax.scatter(x,y)
plt.show()
```

```python
rng=default_rng(18)
size=100

x_azul=rng.uniform(size=size)
y_azul=1.0*x_azul+rng.normal(scale=0.2,size=size)

x_rojo=rng.uniform(size=size)
y_rojo=1.0-1.0*x_rojo+rng.normal(scale=0.2,size=size)

fig,ax=plt.subplots()
ax.scatter(x_azul,y_azul,color="blue")
ax.scatter(x_rojo,y_rojo,color="red")
plt.show()
```

```python
linear= lambda x: x-0.5
quadratic= lambda x: (x-0.25)*(x-0.75)
cubic= lambda x: (x-0.33)*(x-0.5)*(x-0.66)

funciones= [linear,quadratic,cubic]

x=np.linspace(0,1, num=200)

fig, axs= plt.subplots(1,3, figsize=(10,4))
for f,ax in zip(funciones, axs.flatten()):
    ax.plot(x, f(x))
    
plt.show
```

```python
rng=default_rng(24)

fig, axs= plt.subplots(1,3, figsize=(10,4))
for f,ax in zip(funciones, axs.flatten()):
    rng.shuffle(x)
    ax.plot(x, f(x))
    
plt.show()
```

```python
rng=default_rng(25)

c_puntos1=(-1,1)
size1=50

c_puntos2=(1,-1)
size=50

corte=0
pend=1
rango=(-4,4)

fig,ax=plt.subplots()

x1=rng.normal(loc=c_puntos1[0], size=size)
y1=rng.normal(loc=c_puntos1[1], size=size)

x2=rng.normal(loc=c_puntos2[0], size=size)
y2=rng.normal(loc=c_puntos2[1], size=size)

ax.scatter(x1,y1, color='green')
ax.scatter(x2,y2, color='red')

x_linspace=np.linspace(rango[0],rango[1], num=20)
ax.plot(x_linspace,pend*x_linspace+corte, linewidth=25)

plt.show()
```
