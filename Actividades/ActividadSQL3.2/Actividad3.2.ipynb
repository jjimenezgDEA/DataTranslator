## 4.1
### Obtener el “department ID” y el total de salarios a pagar en cada departamento.

## 4.2
### Obtener el salario medio de cada ID de puesto, excluyendo “Programmer”.

```python
sql = """
SELECT avg(salary) AS Avg_Salary
FROM employees AS TS
INNER JOIN jobs AS TA
ON TS.job_id == TA.job_id WHERE job_title NOT LIKE 'Programmer' 
"""
pd.read_sql_query(sql, conn)
```

## 4.3
### Obtener el job ID y el salario máximo de los empleados que su salario máximo es mayor o igual a 4000.

```python
sql = """
SELECT jobs.job_id, max(em.salary) FROM employees em
INNER JOIN
jobs ON em.job_id = jobs.job_id
GROUP BY jobs.job_id
HAVING max(em.salary) >= 4000
"""
pd.read_sql_query(sql, conn)
```

## 4.4
### Obtener el salario medio de todos los departamentos que emplean a más de 10 trabajadores.

```python
sql = """
SELECT department_id, avg(salary) AS AvgSalary FROM employees
GROUP BY department_id
HAVING count(employee_id) > 10
"""
pd.read_sql_query(sql, conn)
```

## 4.5
### Obtener nombre, apellido y  salario de los empleados que tienen un salario más alto que el empleado cuyo apellido es “Bull”.

```python
sql = """
SELECT em.first_name, em.last_name, salary FROM employees em
WHERE salary > (SELECT salary FROM employees em WHERE em.last_name = 'Bull')
"""
pd.read_sql_query(sql, conn)
```

## 4.6
### Obtener nombre y apellido de todos los empleados que trabajan en el departamento de IT.

## 4.7
### Obtener nombre y apellido de los empleados que tienen un “manager” que trabaja en un departamento con sede en EUA.

```python
sql = """
SELECT first_name, last_name
FROM employees a 
INNER JOIN (
SELECT manager_id, country_name
FROM departments AS TA
INNER JOIN locations AS TB
ON TA.location_id == TB.location_id 
INNER JOIN countries AS TC
ON TC.country_id == TB.country_id 
WHERE country_name LIKE '%United States%' AND manager_id IN (200,114,121,103,100,108,205)
) b 
ON a.manager_id==b.manager_id
"""
pd.read_sql_query(sql, conn)
```

## 4.8
### Obtener nombre y apellido de los empleados que son “managers”.

```python
sql = """
SELECT first_name, last_name, job_title
FROM employees AS E
INNER JOIN jobs AS F
ON F.job_id==E.job_id 
WHERE job_title LIKE '%manager%'
"""
pd.read_sql_query(sql, conn)
```

## 4.9
### Obtener nombre, apellido y salario de los empleados cuyo salario es mayor que el salario medio.

```python
sql = """
SELECT em.first_name, em.last_name, salary, en.AVG_Salary, en.job_id FROM employees em
INNER JOIN (select avg(salary) as AVG_Salary, job_id from employees group by job_id) en
ON em.job_id==en.job_id
WHERE salary > en.AVG_Salary
"""
pd.read_sql_query(sql, conn)
```

# Part 5
## 5.1
### Obtener nombre, apellido y salario de los empleados cuyo salario es igual al salario mínimo para su rol.

```python
sql="""
SELECT em.first_name, em.last_name, salary, en.MIN_Salary, en.MAX_Salary, en.job_id FROM employees em
INNER JOIN (select min(salary) AS MIN_Salary, max(salary) AS MAX_Salary, job_id FROM employees GROUP BY job_id) en
ON em.job_id==en.job_id
WHERE salary == en.MIN_Salary
"""
pd.read_sql_query(sql, conn)
```

## 5.2
### Obtener nombre, apellido y salario de los empleados que ganan más que el salario medio y que trabajan en cualquier área de IT.

```python
sql = """
SELECT em.first_name||' '||em.last_name AS Names, em.salary, Salario_Promedio FROM employees em
INNER JOIN departments dep ON em.department_id = dep.department_id
INNER JOIN (select employee_id, job_id, avg(salary) AS Salario_Promedio FROM employees GROUP BY job_id) AS Subquery
ON Subquery.job_id = em.job_id
WHERE dep.depart_name = 'IT' AND salary > Salario_Promedio
"""
pd.read_sql_query(sql, conn)
```

## 5.3
### Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.

```python
sql = """
SELECT em.first_name, em.last_name, salary FROM employees em
WHERE salary > (SELECT salary FROM employees em WHERE em.last_name = 'Bell')
"""
pd.read_sql_query(sql, conn)
```

## 5.4
### Obtener nombre y apellido de los empleados que no son “supervisors”.

```python
sql = """
SELECT first_name, last_name, job_title
FROM employees AS E
INNER JOIN jobs AS F
ON F.job_id==E.job_id 
WHERE job_title NOT LIKE '%supervisors%'
"""
pd.read_sql_query(sql, conn)
```

## 5.5
### Obtener el employee ID, nombre, apellido y salario de todos los empleados cuyo salario es superior a la media de sus departamentos. 

```python
sql = """
SELECT em.employee_id, em.first_name, em.last_name, em.salary, Salario_Promedio FROM employees em
INNER JOIN departments dep ON em.department_id = dep.department_id
INNER JOIN (select employee_id, department_id, avg(salary) AS Salario_Promedio FROM employees GROUP BY department_id) AS Subquery
ON Subquery.department_id = em.department_id
WHERE em.salary > Salario_Promedio
"""
pd.read_sql_query(sql, conn)
```

## 5.7
### Obtener las direcciones (location_id, street_address, city, state_province, country_name) de todos los departamentos.
