## 5.1

```python
meca=stats.poisson(2)
```

```python
meca.pmf(0)
```

```python
plt.bar((range(0,10)),meca.pmf(range(0,10)))
plt.plot((range(0,10)),meca.pmf(range(0,10)),marker='*',c='r')
```

```python
plt.bar((range(0,10)),meca.cdf(range(0,10)))
plt.plot((range(0,10)),meca.cdf(range(0,10)),marker='*',c='r')
```
