## Gr√°fico de l√≠nea

La instrucci√≥n `ax.plot()` entrega un gr√°fico de l√≠nea por defecto.

```python
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

fig, ax = plt.subplots()
ax.plot(x_data, y_data);
```

El gr√°fico muestra los pares (1,1), (2,2), (3,3), (4,4), y (5,5) unidos por una l√≠nea. 

En este caso no es posible observar los puntos.

Cuando los datos son suficientemente diferentes es posible observar el cambio del √°ngulo en la l√≠nea que une cada punto.

```python
x = [1, 2, 3, 4.5, 5]
y = [1, 2.5, 3, 4, 5]

fig, ax = plt.subplots()
ax.plot(x, y);
```

**El orden de los pares tiene importancia**

```python
#same points, different ordering
x = [3, 1, 5, 2, 4.5]
y = [3, 1, 5, 2.5, 4]

fig, ax = plt.subplots()
ax.plot(x, y);
```

## `subplots()` Avanzado

`plt.subplots(n,m)` permite especificar una malla de `n` filas por `m` columnas, con un conjunto de ejes en cada mallado. Los conjuntos de ejes se muestran como un array de numpy.

```python
fig, axs = plt.subplots(2,4)
fig.tight_layout()
```

```python
type(axs[0])
```

```python
axs[0,2]
```

Para graficar en uno de los conjuntos de ejes se puede utilizar su √≠ndice en el array del mallado y despu√©s graficar en este espacio.

```python
fig, axs = plt.subplots(2,4)
axs[0,2].plot(x_sin,y_sin)
fig.tight_layout()
```

#### Tama√±o de subplots
El tama√±o de los subplots tambi√©n puede configurarse con la instrucci√≥n `figsize` dentro de `plt.subplots()`.

```python
fig, axs = plt.subplots(2,4, figsize=(10,4))

axs[0,2].plot(x_sin,y_sin)
axs[0,2].set_title('cool')

fig.tight_layout()
```

### Comunmente se utiliza `ax` para un solo conjunto de ejes, y se utiliza `axs` para m√°s de un conjunto de ejes.

#### Iteraciones entre conjuntos de ejes en un subplot

Cuando se tiene un gran mallado de conjuntos de ejes, indizar manualmente cada uno de ellos se vuelve complicado, por lo que es mas sencillo utilizar un ciclo.

Solo necesitamos escribir `for ax in axs`, aunque al iterar entre un np.array de `n` por `m` el resultado es una _fila_ a la vez, m√°s no un _elemento_. Es por esto que se tiene que utilizar la instrucci√≥n `.flatten()` para transformar el array a una sola dimensi√≥n.

```python
axs.flatten().shape
```

```python
fig, axs = plt.subplots(2,4, figsize=(10,4))

for i, ax in enumerate(axs.flatten()):
    ax.scatter(x_sin,y_sin*(i+1))
    ax.set_title('Gr√°fico n√∫mero {}'.format(i+1))

fig.tight_layout()
```

## plt.plot

Existe una manera m√°s veloz de crear gr√°ficos.

```python
plt.scatter(x_data, y_data)
```

And MANY online examples will use this method to plot.

```python
#this is OK

plt.scatter(x_data, y_data)
plt.title('Line')
plt.xlabel('x')
plt.ylabel('y')
```

Sin embargo, este m√©todo se vuelve demasiado complicado en el momento en que se a√±ade m√°s que los t√≠tulos.

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
plt.title('Seno y Seno**2')

fig1 = plt.gca()

ax2 = fig.add_subplot(1,2,2)
plt.title('L√≠nea')

fig2 = plt.gca()

ax1_stuff = ax1.plot(x_sin, y_sin, color='m', alpha=.25)

ax1.plot(x_sin, y_sin**2)

# ax1_stuff[0].set_alpha(.25) # M√©todo para utilizar el par√°metro alpha

ax2.plot(x_sin, x_sin)

line = ax1_stuff[0]
line.set_alpha(.55)

fig.suptitle('Dos l√≠neas', fontsize=44, y=1.1)
fig.tight_layout()
```

Por esto se recomienda utilizar `fig, axs = plt.subplots()`

# Escribir funciones para crear gr√°ficos

**Siempre que tenga el instinto de copiar y pegar es se√±al de que necesitas escribir una funci√≥n.**

La mejor manera de escribir una funci√≥n con gr√°ficos es hacer una funci√≥n que tome un eje y algunos datos y los grafique en ese eje.

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
fig, ax = plt.subplots()
out = my_plotter(ax, x_sin, y_sin, {'linewidth':3, 'linestyle':'-.'})
```

### Actividad
Definir una funci√≥n para graficar.

```python
# C√≥digo
def scatterORline(ax,plt_type=0,size=50,color='blue',centro=(0,0),rango=(-5,5),mb=(1,0)):
    from numpy.random import default_rng
    rng=default_rng(42)
    if plt_type==0:
        x=rng.normal(loc=centro[0],size=size)
        y=rng.normal(loc=centro[1],size=size)
        ax.scatter(x,y,color=color)
    else:
        x=np.linspace(rango[0],rango[1],num=size)
        y=mb[0]*x+mb[1]
        ax.plot(x,y,color=color)
        
fig,ax=plt.subplots()
scatterORline(ax)
scatterORline(ax,color='red',centro=(1,-1))
scatterORline(ax,1,rango=(-2,2),mb=(3,-1),color='green',centro=(1,-1))
```
